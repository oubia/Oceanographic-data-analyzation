```python
def biom_espèce(ch):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    biomasse_esp = pd.read_csv('biomasse(espèce).csv',sep=';',header=0)
    dic=biomasse_esp.to_dict()
    key=list(dic.keys())
    
    def moyenne(ch):
        k=' '
        for i in range(len(ch)):
            if ch[i] in '-_,;':
                k=ch [i]
                
        if k in '-_,;':
            ch=ch.split(k)
            m=(float(ch[0])+float(ch[1]))/2
            return m
        else:
            return float(ch)
    

    for i in key[1:]:
        for j in range(len(dic[key[0]])):
            dic[i][j]=moyenne(dic[i][j])
            
    def get_index(ch) :
        y=[]
        for i in range(len(dic[key[0]])):
            if dic[key[0]][i]==ch:
                y.append(i)
        return y
    k=get_index('Coryphaena hippurus')
    k=k[0]
    k
    biom=[]
    year=[]
    for i in key[1:]:
        biom.append(dic[i][k])
        year.append(int(i))
        
    plt.figure(figsize=(10,8))
    plt.plot(year,biom,c='red',label=ch)
    plt.title("analyse de biomasse")
    plt.xlabel("Années")
    plt.ylabel("Biomasse (tonnes/km²)")
    plt.legend()
    plt.show()

    
```

```python
biom_espèce('Coryphaena hippurus')
```
